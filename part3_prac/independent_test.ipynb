```python
import pandas as pd
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
print(df.head(2))
```

```python
# 1. 검정 통계량 구하기
df.describe()
```

```python
# 2. p-value 구하기
from scipy import stats

a = df[df['충전기'] == 'New']
b = df[df['충전기'] == 'Old']

statistics, pvalue = stats.ttest_ind(a['충전시간'], b['충전시간'], alternative='less', equal_var=True)
print("{:.10f}".format(pvalue))
```
