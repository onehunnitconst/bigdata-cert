```python
import pandas as pd

df = pd.DataFrame({
    '구분': ['1회', '2회', '3회', '4회', '5회'],
    '경험수': [550, 250, 100, 70, 30],
    '전국경향': [0.6, 0.25, 0.08, 0.05, 0.02], 
})

df.head()
```

```python
df['경험수'][4] / 1000
```

```python
from scipy.stats import chisquare

statistics, pvalue = chisquare(df['경험수'], df['전국경향'] * 1000)
```
