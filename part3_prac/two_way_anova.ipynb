```python
import pandas as pd
df = pd.read_csv("./data/tomato2.csv")
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('수확량~C(비료유형)*C(물주기)', data=df).fit()
anova_table = anova_lm(model)
anova_table
```
