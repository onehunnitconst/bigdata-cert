```python
import pandas as pd
df = pd.read_csv("./data/math.csv")
print(df.head())
```

```python
# 1. 각 그룹의 수학성적에 대해 샤피로-윌크 검정을 통한 정규성 확인
from scipy import stats

groups = ['group_A', 'group_B', 'group_C', 'group_D']

for group in groups:
    df_g = df[df['groups'] == group]
    statistics, pvalue = stats.shapiro(df_g['scores'])
    print(f'{group}:', '{:.5f}'.format(pvalue))
```

0.05 이상이므로 귀무가설(정규성 가지고 있다) 채택

```python
scores = [ df[df['groups'] == group]['scores'] for group in groups ]

statistics, pvalue = stats.levene(scores[0], scores[1], scores[2], scores[3])
print('{:.5f}'.format(pvalue))
```

0.05 이상이므로 귀무가설(등분산성을 가지고 있다) 채택

```python
import statsmodels as sm
from statsmodels.formula.api import ols

model = ols('scores~groups', df).fit()
```

```python
from statsmodels.stats.anova import anova_lm

print(anova_lm(model))
```
