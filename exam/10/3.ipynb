```python
import pandas as pd

df = pd.read_csv('./hamspam.csv')
```

```python
df['words'] = df['text'].str.split()
df.head()
```

```python
df['words_count'] = df['words'].str.len()
df.head()
```

```python
msg_mean = df.groupby('label')['words_count'].mean()
```

```python
abs(round(msg_mean['spam'] - msg_mean['ham'], 3))
```
