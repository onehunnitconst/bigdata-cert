```python
import pandas as pd

df = pd.read_csv('attrition.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from statsmodels.formula.api import logit

formula = 'attrition ~ age + income + overtime'
model = logit(formula, df).fit()
model.summary()

# -0.002
```

```python
test = pd.DataFrame({
    'age': [40],
    'income': [4500],
    'overtime': [1]
})

test.head()
```

```python
import numpy as np

round(np.exp(model.params['age']), 3)
```

```python
round(model.predict(test)[0], 3)
```
