```python
import pandas as pd

df = pd.read_csv('heating.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from statsmodels.formula.api import ols

formula = 'heating_load ~ wall + roof + glazing + height'
model = ols(formula, df).fit()
model.summary()
```

```python
round(model.params[1:].sum(), 3)
```

```python
formula = 'heating_load ~ roof + glazing + height'
model = ols(formula, df).fit()
model.summary()
```

```python
round(model.rsquared, 3)
```

```python
test = pd.DataFrame({
    'wall': [20],
    'roof': [150],
    'glazing': [20],
    'height': [5]
})

round(model.predict(test)[0], 3)
```
