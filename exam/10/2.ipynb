```python
import pandas as pd

df = pd.read_csv('cafe_sales.csv')
```

```python
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df['order_month'] = df['order_date'].dt.strftime('%Y%m')
df.head()
```

```python
sales = df.groupby('order_month')['price'].sum().sort_values(ascending=False)
sales.iloc[1]
```

```python
idx = sales.index[3]
idx
```

```python
cond = df['order_month'] == '202410'
cat_sales = df[cond].groupby('category')
cat_sales['price'].sum().max()
```
