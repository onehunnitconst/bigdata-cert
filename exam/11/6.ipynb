## 문제 5
제공된 데이터(`promotion_data.csv`)는 기업의 마케팅 홍보 활동에 따른 매출 데이터로, 방문 횟수, 웹페이지 방문 시간, 광고 클릭 수, 홍보 예산 등에 따른 매출액(sales)에 관한 자료이다. 이 데이 터를 이용하여 통계 분석을 수행하고, 제시된 소문항에 답하시오.

> 제공 데이터 (promotion_data.csv)
> - 종속 변수 (타겟 변수): sales, 매출액
> - 독립 변수 (설명 변수):
> - VISI_count: 방문 횟수
> - page_time: 웹페이지 방문 시간(초)
> - ad_clicks: 광고 클릭 수
> - promotion_budget: 홍보 예산
> ```python
> import pandas as pd
> df = pd.read_csv("promotion_data.csv")
> ```

1. 홍보 전후 매출 평균이 35,000원과 같은지 단일 t검정(t-lest)을 수행하고, p-value를 구하시오. (반올
림하여 소수 셋째 자리까지 작성)
2. 매출액(sales)과 다른 모든 변수들 간의 피어슨 상관계수를 구하고, 그 중 상관관계가 가장 강한(절댓
값이 가장 큰) 상관계수 값을 구하시오. (반올림하여 소수 셋째 자리까지 작성)
3. 매출액(sales)을 종속변수로, 나머지 모든 변수를 독립변수로 하여 회귀분석을 수행하시오. 절편을 포
함한 모든 변수 중에서 p-value가 가장 작은 변수(통계적으로 가장 유의한 변수)의 회귀계수를 구하시오. (반올림 하여 소수 셋째 자리까지 작성)

````python
import pandas as pd
df = pd.read_csv("promotion_data.csv")
````

````python
df.head()
````

````python
from scipy.stats import ttest_1samp

statistics, pvalue = ttest_1samp(df['sales'], 35000)

pvalue > 0.05
````

````python
from scipy.stats import pearsonr

# pearsonr()
df.corr(method='pearson')

# 0.575396
````

````python
from statsmodels.formula.api import ols

formula = 'sales ~ visit_count + page_time + ad_clicks + promotion_budget'
model = ols(formula, df).fit()
model.summary()

# p-value가 가장 작은 변수는 page_time
# 회귀계수는 130.7392
````
