## 문제 1
- 1-1. 2001년부터 2021년까지 각 온도별로 온실가스 배출량이 가장 많은 국가를 찾고, 그 중 1위를 가장 많이 차지한 국가의 1위 횟수를 정수로 구하시오
- 1-2. 다음 공식으로 각 국가의 CAGR(연평균 성장률)을 계산하고, 그 중 최댓값을 소수점 셋째자리까지 구하시오

```python
import pandas as pd

df = pd.read_csv('greenhouse_gas.csv')
df.head()
```

```python
cols = df.columns[1:]

max = df[cols].idxmax()
max

df.loc[max]['Country'].value_counts()
```

```python
df['CAGR'] = (df['2021'] / df['2001']) ** (1/20) - 1
```

```python
round(df['CAGR'].max(), 3)
```
