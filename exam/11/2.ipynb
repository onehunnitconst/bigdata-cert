## 문제 2
1. 결측치가 가장 많은 컬럼 찾기
2. 해당 컬럼의 결측치를 중앙값으로 대체
3. 결측치를 대체한 후 해당 컬럼의 평균을 소수점 셋째 자리까지 구하기

```python
import pandas as pd

df = pd.read_csv('sensor_data.csv')
df.head()
```

```python
most_na = df.isnull().sum().idxmax()
```

```python
df.isnull().sum()
```

```python
most_na_median = df[most_na].median()
df[most_na] = df[most_na].fillna(most_na_median)
```

```python
df.isnull().sum()
```

```python
round(df[most_na].mean(), 3)
```
