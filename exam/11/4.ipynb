## 문제 4
제공된 학습용 데이터(train.csv)는 온라인 게임 플레이어의 활동 데이터이다. 학습용 데이터
를 활용하여 유저 등급(User_Level)을 예측하는 모형을 개발하고, 모형을 평가용 데이터(test.csv)에 적용하여 유저 등급을 예측하시오. 예측 결과는 아래의 [유의 사항]과 [제출 형식]을 준수하여, CSV 파 일(result.csv)을 생성하는 코드를 제출하시오.

### 유의 사항
1. 예측 결과는 F1 Score(macro) 평가 지표에 따라 평가함
2. 타겟 변수(User_Level)는 0, 1, 2 총 3개의 클래스로 구성됨 (O: Low, 1: Medium, 2: High)

### 제출 형식
1. CSV 파일명 : result.csv
2. 예측 유저 등급 컬럼명: pred
3. 제출 컬럼 개수: pred 컬럼 1개
4. 평가용 데이터 개수와 예측 결과 데이터 개수 일치: 903개

### 제공 데이터
- train.csv: 학습용 데이터, 약 2,614개
- test.csv: 평가용 데이터, 약 903개

### CSV 파일 형식 및 확인 방법
• CSV 파일명 : result.csv
• 예측 유저 등급 컬럼명 : pred

````python
target = train.pop('User_Level')
````

````python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
print(X_train.shape, X_val.shape)
print(y_train.shape, y_val.shape)
````

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_val)
````

````python
from lightgbm import LGBMClassifier

lg = LGBMClassifier(random_state=0, verbose=-1)
lg.fit(X_train, y_train)
lg_pred = lg.predict(X_val)
````

````python
from sklearn.metrics import f1_score

rf_score = f1_score(y_val, rf_pred, average='macro')
rf_score
````

````python
lg_score = f1_score(y_val, lg_pred, average='macro')
lg_score
````

````python
pred = lg.predict(test)

result = pd.DataFrame({ 'Pred': pred })
result.to_csv('result.csv', index=False)
````
