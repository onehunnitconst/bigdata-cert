```python
import pandas as pd

df = pd.read_csv('retention.csv')
```

```python
df.head()
```

```python
from statsmodels.formula.api import logit

formula = 'Churn ~ MonthlyCharges + CustomerTenure + HasPhoneService + HasTechInsurance'
model = logit(formula, df).fit()
```

```python
model.summary()
```

```python
round(model.pvalues['MonthlyCharges'], 3)
```

```python
import numpy as np

round(np.exp(model.params['HasPhoneService']), 3)
```
