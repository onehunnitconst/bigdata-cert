```python
import pandas as pd

df = pd.read_csv("design.csv")

df.head()
```

```python
train = df[:140]
test  = df[140:]
```

```python
from statsmodels.formula.api import ols

formula = "design ~ c1 + c2 + c3 + c4 + c5"
model = ols(formula, train).fit()
model.summary()

# 0.05보다 작은 변수: 3개
```

```python
formula = "design ~ c1 + c2 + c4"
model = ols(formula, train).fit()
model.summary()
```

```python
pred = model.predict(train)
```

```python
from scipy.stats import pearsonr

round(pearsonr(train['design'], pred).statistic, 3)
```

```python
from sklearn.metrics import root_mean_squared_error

pred = model.predict(test)
round(root_mean_squared_error(test['design'], pred), 3)
```
