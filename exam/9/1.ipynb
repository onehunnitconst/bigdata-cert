```python
import pandas as pd

df = pd.read_csv('loan.csv')

df.head()
```

```python
df_g = df.groupby(by=['지역코드','성별']).sum()
df_g['총대출액'] = df_g['신용대출'] + df_g['담보대출']
df_g.head()
```

```python
df_g = df_g.unstack()
```

```python
df_g['차이'] = abs(df_g['총대출액'][1] - df_g['총대출액'][2])
df_g = df_g.sort_values(by='차이', ascending=False)
df_g['차이'].idxmax()
```
