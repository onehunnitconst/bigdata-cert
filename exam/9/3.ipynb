```python
import pandas as pd

df = pd.read_csv('hr.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['만족도'] = df['만족도'].fillna(df['만족도'].mean())
df.isnull().sum()
```

```python
fillna_value = df.groupby(['부서', '성과등급'])['근속연수'].transform("mean").astype(int)
df['근속연수'] = df['근속연수'].fillna(fillna_value)
df.isnull().sum()
```

```python
df['연봉/근속연수'] = df['연봉'] / df['근속연수']
df['연봉/만족도'] = df['연봉'] / df['만족도']
df.head()
```

```python
a = df.sort_values('연봉/근속연수', ascending=False).iloc[2]['근속연수']
```

```python
b = df.sort_values('연봉/만족도', ascending=False).iloc[1]['교육참가횟수']
```
