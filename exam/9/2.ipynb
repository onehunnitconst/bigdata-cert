```python
import pandas as pd

df = pd.read_csv('crime.csv')
df.head()
```

```python
df_a = pd.melt(df, id_vars=["연도", "구분"], var_name="범죄유형", value_name="건수")
```

```python
df_a.head()
```

```python
df_b = df_a.pivot_table(index=['연도', '범죄유형'], columns='구분', values='건수').reset_index()
```

```python
df_b.head()
```

```python
df_b['검거율'] = df_b['검거건수'] / df_b['발생건수']

df_b.head()
```

```python
max_crime = df_b.groupby('연도')['검거율'].idxmax()
```

```python
df_b.loc[max_crime]['검거건수'].sum()
```
