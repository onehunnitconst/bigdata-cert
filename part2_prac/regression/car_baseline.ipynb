```python
import pandas as pd

train = pd.read_csv("./data/car_train.csv")
test  = pd.read_csv("./data/car_test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
cols = train.select_dtypes(include='object').columns

for col in cols:
    set_train = set(train[col])
    set_test = set(test[col])

    same = set_train == set_test

    print(f"{col}: ", same)
```

```python
target = train.pop('Price')

from sklearn.preprocessing import LabelEncoder

combined = pd.concat([train, test])
cols = train.select_dtypes(include='object')

for col in cols:
    le = LabelEncoder()
    combined[col] = le.fit_transform(combined[col])

n_train = len(train)
train = combined[:n_train]
test = combined[n_train:]
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import root_mean_squared_log_error

result = root_mean_squared_log_error(y_val, pred)
print("score: ", result)
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({"pred": pred})
submit.to_csv("./submit/car_baseline_result.csv", index=False)
```
