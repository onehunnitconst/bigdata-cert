```python
import pandas as pd

train = pd.read_csv("./data/glass_train.csv")
test  = pd.read_csv("./data/glass_test.csv")
```

```python
target = train.pop('Type')
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_val, pred, average='weighted')
print('f1_weighted: ', score)
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred': pred})
submit.to_csv("./submit/glass_baseline_result.csv", index=False)
```
